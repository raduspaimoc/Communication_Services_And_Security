```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_prev = pd.read_csv("without_traffic.csv")
```

```python
df_prev
```

```python
plt.figure(figsize=(12,7))
#plt.figure(figsize=(18, 10), dpi=80, facecolor='white', edgecolor='k')
#pl.plot(time[0], rate[0], linewidth=1, color='blue', label='Without Traffic Shaping')
#pl.plot(time[1], rate[1], linewidth=1, color='red', label='Traffic Shaping rate 200 kB/s')
#pl.plot(time[2], rate[2], linewidth=1, color='green', label='Traffic Shaping rate 800 kB/s')
plt.plot(df_prev['time'], df_prev['rate'], linewidth=1, label='Traffic Shaping rate 1000 kB/s')


plt.legend()
plt.ylabel('rate Bytes/second')
plt.xlabel('Seconds')
plt.yticks(np.arange(0,230000,10000))
plt.xticks(np.arange(0,180,10))
plt.grid()
#pl.savefig("traffic2.png")
plt.show()
```

```python
df_1000 = pd.read_csv("avg_1000.csv")
df_800 = pd.read_csv("avg_800.csv")
df_no_shapping = pd.read_csv("avg_no_shapping.csv")
```

```python
plt.figure(figsize=(20,10))
#plt.figure(figsize=(18, 10), dpi=80, facecolor='white', edgecolor='k')
plt.plot(df_no_shapping['time'], df_no_shapping['rate'], linewidth=2, label='Without Traffic Shaping')
plt.plot(df_800['time'], df_800['rate'], linewidth=2, label='Traffic Shaping rate 800 kB/s')
plt.plot(df_1000['time'], df_1000['rate'], linewidth=2, label='Traffic Shaping rate 1000 kB/s')


plt.legend()
plt.ylabel('rate Bytes/second')
plt.xlabel('Seconds')
plt.yticks(np.arange(0,230000,10000))
plt.xticks(np.arange(0,200,10))
plt.grid()
#pl.savefig("traffic2.png")
plt.show()
```
