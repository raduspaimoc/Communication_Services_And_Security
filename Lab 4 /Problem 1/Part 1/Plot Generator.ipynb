# Plot Generator

This script generates the plots for the previous generated .csv files. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_1000 = pd.read_csv("avg_1000.csv")
df_800 = pd.read_csv("avg_800.csv")
df_no_shapping = pd.read_csv("avg_no_shapping.csv")
```

```python
plt.figure(figsize=(20,10))
#plt.figure(figsize=(18, 10), dpi=80, facecolor='white', edgecolor='k')
plt.plot(df_no_shapping['time'], df_no_shapping['rate'], linewidth=2, label='No Traffic Shaping')
#plt.plot(df_800['time'], df_800['rate'], linewidth=2, label='Traffic Shaping rate 800 kB/s')
plt.plot(df_1000['time'], df_1000['rate'], linewidth=2, label='Traffic Shaping rate 1000 kB/s')


plt.legend()
plt.ylabel('Rate (Bytes/second)')
plt.xlabel('Seconds')
plt.yticks(np.arange(0,250000,20000))
plt.xticks(np.arange(0,200,10))
plt.grid()
plt.savefig("plt_100.png")
plt.show()
```
